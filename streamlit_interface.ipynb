```python
%%writefile utils.py

import os
import re
import torch
import faiss
import requests
import pdfplumber
import numpy as np
import pandas as pd
import pytesseract
import faster_whisper
import streamlit as st
from bs4 import BeautifulSoup
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi
from PIL import Image
from docx import Document
from sentence_transformers import SentenceTransformer
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import mimetypes
import nltk
from nltk.tokenize import sent_tokenize
from moviepy.editor import VideoFileClip
import whisper
import tempfile

# ✅ Streamlit Config (Must be at the top)
st.set_page_config(page_title="RAG Chatbot", layout="wide")

# ✅ Download necessary NLTK models
nltk.download("punkt")

# ✅ Load & Cache Models
@st.cache_resource()
def load_models():
    print("🔹 Loading Models...")
    summarization_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn").to("cuda" if torch.cuda.is_available() else "cpu")
    summarization_tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
    embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
    whisper_model = whisper.load_model("base")  # ✅ Whisper model cached
    return summarization_model, summarization_tokenizer, embedding_model, whisper_model

summarization_model, summarization_tokenizer, embedding_model, whisper_model = load_models()

dimension = 384  # MiniLM embedding size
faiss_index = faiss.IndexFlatL2(dimension)
documents = []  # Store extracted sentences

# ✅ Web & YouTube Extraction
@st.cache_data()
def extract_text_from_web(url):
    try:
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}, timeout=10)
        soup = BeautifulSoup(response.text, 'html.parser')
        paragraphs = soup.find_all('p')
        return "\n".join([para.get_text() for para in paragraphs]).strip()
    except:
        return None

@st.cache_data()
def extract_text_from_youtube(url):
    try:
        video_id = url.split("v=")[-1]
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        return " ".join([t['text'] for t in transcript])
    except:
        return None

# ✅ File Extraction Functions
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += (page.extract_text() or "") + "\n"
    return text.strip()

def extract_text_from_docx(docx_path):
    return "\n".join([para.text for para in Document(docx_path).paragraphs]).strip()

def extract_text_from_image(img_path):
    return pytesseract.image_to_string(Image.open(img_path)).strip()

def extract_text_from_video(video_path):
    with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as audio_temp:
        video = VideoFileClip(video_path)
        video.audio.write_audiofile(audio_temp.name)
        result = whisper_model.transcribe(audio_temp.name)
        os.remove(audio_temp.name)
    return result['text'].strip()

# ✅ Main Extraction Function (Handles Multiple File Types)
def extract_text(file_path):
    mime_type, _ = mimetypes.guess_type(file_path)
    print(f"🔹 MIME Type Detected: {mime_type}")  # Debugging print

    # ✅ Fix: Handle NoneType
    if mime_type is None:
        return "❌ Unsupported file type."

    if "pdf" in mime_type:
        return extract_text_from_pdf(file_path)
    elif "officedocument" in mime_type or file_path.endswith(".docx"):
        return extract_text_from_docx(file_path)
    elif "image" in mime_type:
        return extract_text_from_image(file_path)
    elif "video" in mime_type:
        return extract_text_from_video(file_path)

    return "❌ Unsupported file type."

# ✅ Add Document to FAISS Index
def add_document(text):
    global faiss_index, documents
    sentences = sent_tokenize(text)
    embeddings = embedding_model.encode(sentences, convert_to_numpy=True)
    faiss_index.add(embeddings)
    documents.extend(sentences)

# ✅ Retrieve Top k Sentences
def retrieve_top_k(query, k=3):
    if not documents:
        return "❌ No documents available."
    query_embedding = embedding_model.encode([query], convert_to_numpy=True)
    _, indices = faiss_index.search(query_embedding, k)
    return "\n".join([documents[idx] for idx in indices[0] if idx < len(documents)])

# ✅ Generate Summary
def generate_summary(text):
    inputs = summarization_tokenizer("summarize: " + text, return_tensors="pt", max_length=512, truncation=True).to("cuda" if torch.cuda.is_available() else "cpu")
    with torch.no_grad():
        summary_ids = summarization_model.generate(inputs.input_ids, max_length=150, min_length=50, num_beams=4, early_stopping=True)
    return summarization_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# 🎯 *Streamlit UI*
st.title("📄💬 RAG Chatbot with File & Web Processing")

uploaded_files = st.file_uploader("Upload files (PDF, DOCX, CSV, Images, Videos)", accept_multiple_files=True)
url_input = st.text_input("Enter a webpage or YouTube URL")

if st.button("Process Input"):
    with st.spinner("Processing..."):
        if url_input:
            extracted_text = extract_text_from_youtube(url_input) if "youtube" in url_input else extract_text_from_web(url_input)
            if extracted_text:
                add_document(extracted_text)
                st.success("✅ URL processed successfully!")

        for file in uploaded_files:
            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file.write(file.getbuffer())
                temp_file.flush()  # ✅ Ensure file is fully written before processing

                extracted_text = extract_text(temp_file.name)
                os.remove(temp_file.name)  # ✅ Remove temp file after processing

                if extracted_text:
                    add_document(extracted_text)
                    st.success(f"✅ {file.name} processed successfully!")
                else:
                    st.error(f"❌ Failed to process {file.name}. Unsupported file type?")

st.divider()
st.subheader("Ask a Question")
query = st.text_input("Enter your query")
if query:
    retrieved_text = retrieve_top_k(query)
    summary = generate_summary(retrieved_text)
    st.write("📌 *Top Retrieved Contexts:*")
    st.info(retrieved_text)
    st.write("📌 *Abstractive Summary:*")
    st.success(summary)
```
